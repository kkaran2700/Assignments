```python
2. Given a recurrence relation, solve it using the substitution method:
a. T(n) = T(n-1) + c
b. T(n) = 2T(n/2) + n
c. T(n) = 2T(n/2) + c
d. T(n) = T(n/2) + c
```

```python
2(a). T(n) = T(n-1) + c

T(n) = T(n-1) + c ----->>>>>(Equation - 1)

T(n-1) = T((n-1)-1) + c
T(n-1) = T(n-2) + c   ----->>>>>(Equation -2)

substituting Eq(1) in Eq(2), we get

T(n) = T(n-2) + c + c
T(n) = T(n-2) +2c  ----->>>>>(Equation-3)

T(n-2) = T((n-2) - 1) + c
T(n-2) = T(n-3) + c ----->>>>>(Equation -4)


substituting Eq(4) in Eq(3), we get
T(n) = T(n-3) + c + 2c
T(n) = T(n-3) + 2c

hence, for k terms we can write

T(n) = T(n-k) + k.c ----->>>>>EQUATION(5)

n-k = 1
k = n-1

substitue k = n-1 in Eq(5), we get
T(n) = T(n- n + 1) + (n-1)c
T(n) = 1 + nc -1
T(n) = nc

In this case ...we can say Time complexity is O(n)

```

```python
2(b). T(n) = 2T(n/2) + n

T(n) = 2T(n/2) + n ----->>>>>Eq(1)

T(n/2) = 2T(n/2^2) + n/2 ---->>>>>Eq(2)

Substitute Eq(2) in Eq(1)

T(n) = 2(2T(n/2^2) + n/2) + n 
T(n) = 2^2T(n/2^2) +2n ----->>>>>Eq(3)

T(n/2^2) = 2T(n/2^3) + n/2^2 ----->>>>>Eq(4)

Substitute Eq(4) in Eq(3)

T(n) = 4(2T(n/2^3) + n/2^2) + 2n
T(n) = 2^3T(n/2^3) + 3n -----Eq(5)

for k terms we get

T(n) = 2^kT(n/2^k) + kn

n/2^k = 1
n = 2^k
apply log2 on both sides we get
log2 n = log2 2^k
log2 n = k log2 2

k = log2 n

substitute k  we get
= 2^log2 n.T(n/2^log2 n) + (log2 n)n
= n + (log2 n)n

hence time complexity is O(log2 n)
```

```python
2(c). T(n) = 2T(n/2) + c

T(n) = 2T(n/2) + c ----->>>>>Eq(1)

T(n/2) = 2T(n/2^2) + c ----->>>>>Eq(2)

substitute Eq(2) in Eq(1), we get

T(n) = 2(2T(n/2^2) + c) + c
T(n) = 2^2T(n/2^2) + 2c ----->>>>>Eq(3)

T(n/2^2) = 2T(n/2^3) + c ----->>>>>Eq(4)
substitute Eq(4) in Eq(3), we get

T(n) = 2^2(2T(n/2^3) + c) + 2c
T(n) = 2^3T(n/2^3) + 3c ----->>>>>eq(5)

for k terms we can conclude

T(n) = 2^kT(n/2^k) + kc ----->>>>>Eq(6)

n/2^k = 1
n = 2^k
apply log2 on both sides
log2 n = log2 2^k
log2 n = k log2 2 
k = log2 n

subsitute k value in Eq(6)
(2^log2 n)T(n/2^log 2 n) + (log2 n)c
n + (log2 n)c
hence we can conclude that
time complexity = O(log2 n)
```
