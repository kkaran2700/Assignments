```python
# Problem 4: Library Catalog Create classes representing a library and a book. Implement methods to add books to the library, borrow books, and display available books.

class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.available = True
        
    def __str__(self):
        return f"{self.title} by {self.author}"
    
class library():
    def __init__(self):
        self.books = []
        
    def addbook(self,book):
        self.books.append(book)
        
    
        
    def borrowbook(self, title):
        for book in self.books:
            if book.title == title and book.available:
                book.available = False
                return f"Successfully borrowed {book.title}"
        return "Book not available"
    
    def returnbook(self, title):
        for book in self.books:
            if book.title == title and not book.available:
                book.available = True
                return f"Successfully returned {book.title}"
        return "Book not found or already returned"
    
    def display_available_books(self):
        available_books = [book for book in self.books if book.available]
        if available_books:
            return "\n".join(str(book) for book in available_books)
        else:
            return "No books available"
    
    
b1 = Book('maths', 'mathteacher')
b2 = Book('phy', 'phyteacher')
b3 = Book('bio', 'bioteacher')

l = library()
l.addbook(b1)
l.addbook(b2)
l.addbook(b3)


print(l.display_available_books())
print(l.borrowbook('maths'))
print(l.borrowbook('phy'))
print(l.borrowbook('bio'))
print(l.borrowbook('social'))

print(l.display_available_books())

print(l.returnbook('maths'))
print(l.returnbook('phy'))

print(l.display_available_books())
```

```python
# Problem 5: Product Inventory Create classes representing a product and an inventory system. Implement methods to add products to the inventory, update product quantity, and display available products.
class product:
    def __init__(self, productname, qty):
        self.productname = productname
        self.qty = qty

        
    def __str__(self):
        return f"{self.productname} and qunatity is {self.qty}"
    

class inventory:
    def __init__(self):
        self.available_products = []
        
    def addproducts(self, productt):
        found = False
        for item in self.available_products:
            if item.productname == productt.productname:
                item.qty += 1
                found = True
                break
        if not found:
            self.available_products.append(productt)
        
    def displayproducts(self):
        return self.available_products
        
    
i = inventory()

p1 = product("bag", 1)
p2 = product("shoe",1)
p3 = product("belt",1)

print(p1.productname)

print(p1.qty)

print(p2.productname)

print(p2.qty)

print(p3.productname)

print(p3.qty)


i.addproducts(p1)
i.addproducts(p2)
i.addproducts(p3)
i.addproducts(p3)
i.addproducts(p3)

r = i.displayproducts()
for p in r:
    print(p)
```

```python
# Problem 6: Shape Calculation Create a class representing a shape with attributes like length, width, and height. Implement methods to calculate the area and perimeter of the shape.

class cuboid:
    def __init__(self,l,b,h):
        self.l = l
        self.b = b
        self.h = h
        
    def area(self):
        return 2*((self.l*self.b) + (self.l*self.h) + (self.b*self.h))
    
    def perimeter(self):
        return 4*(self.l + self.b + self.h)

c1 = cuboid(3,4,5)
print(c1.area())
print(c1.perimeter())
```

```python
# Problem 7: Student Management Create a class representing a student with attributes like student ID, name, and grades. Implement methods to calculate the average grade and display student details.

class Student:
    def __init__(self, name, ID, grade):
        self.name = name
        self.id = ID
        self.grade = grade
        
    def __add__(self, other):
        combined_name = self.name + " & " + other.name
        combined_grade = (self.grade + other.grade) / 2
        return Student(combined_name, " & ".join([str(self.id), str(other.id)]), combined_grade)

    def average_grade(self):
        return self.grade
    
    def display_details(self):
        print("Name:", self.name)
        print("ID:", self.id)
        print("Grade:", self.grade)

s1 = Student('Karan', 100, 70)
s2 = Student('Kranthi', 101, 80)
s3 = Student('Saikiran', 102, 75)

result = s1 + s2 + s3
result.display_details()
```

```python
# Problem 9: Social Media Profile Create a class representing a social media profile with attributes like username and posts. Implement methods to add posts, display posts, and search for posts by keyword.

class socialmedia:
    def __init__(self):
        self.profdict = dict()
        
        
        
    def set_username(self,username):
        self.username = username
        if username not in self.profdict:
            self.profdict[self.username] = []
        
        
    def set_post(self,username, post):
        self.set_username(username)
        self.profdict[self.username].append(post)
        

    def add_post(self,username, addpost):
        self.set_username(username)
        if addpost in self.profdict[username]:
            print("this post is already posted")
        else:
            self.profdict[self.username].append(addpost)
                    

    def display_posts(self):
        return self.profdict
        
    def search_posts(self, username):
        return self.profdict[self.username]
    
sm1 = socialmedia()

sm1.set_username("karan")
sm1.set_username("kiran")

sm1.set_post('karan', "selfie")
sm1.set_post('kiran',"random")

sm1.add_post("kiran",'random')
sm1.add_post("karan","random")
sm1.add_post('karan','handpose')
print(sm1.display_posts())

sm1.search_posts('karan')

        
```

```python
# Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. Implement methods to add tasks, mark tasks as completed, and display pending tasks.
class toDoList:
    def __init__(self, tasks = None, dueDate = None):
        self.tasks = tasks
        self.dueDate = dueDate
        self.tasklist = dict()
        
    def setTask(self, tasks):
        self.tasks = tasks
        if self.tasks not in self.tasklist:
            self.tasklist[self.tasks] = []
    
    def setdueDate(self,tasks, dueDate):
        # self.dueDate = dueDate
        self.setTask(tasks)
        self.tasklist[self.tasks].append(dueDate)
        
    def taskcompleted(self,tasks,status):
        if status == True:
            self.tasklist[self.tasks].append('completed')
        else:
            self.tasklist[self.tasks].append('need to complete')
            
    def display_pending_tasks(self):
        for i in self.tasklist[self.tasks]:
            if i == 'need to complete':
                return [self.tasklist[self.tasks]]
            
            
    def displayalltasks(self):
        return self.tasklist
    
t = toDoList()
t.setTask('assignment-1')
t.setdueDate('assignment-1','28th Feb 2024')
t.taskcompleted('assignment-1',True)

t.setTask('assignment-2')
t.setdueDate('assignment-2', '29th Feb 2024')
t.taskcompleted('assignment-2', False)

t.displayalltasks()


    
    
    
```
